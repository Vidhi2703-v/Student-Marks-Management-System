```python
def students():
    student_data = {}

    while True:
        print("\n--- Student Marks Management System ---")
        print("1. Add Student")
        print("2. Display All Students")
        print("3. Calculate Average Marks")
        print("4. Find Top Student")
        print("5. Exit")

        try:
            choice = int(input("\nEnter your choice: "))
        except ValueError:
            print("Please enter a valid number!")
            continue

        # Add students
        if choice == 1:
            name = str(input("Enter student name: "))
            try:
                marks = int(input("Enter marks: "))
                if name:
                    student_data[name] = marks
                    print("Student added successfully!")
                else:
                    print("Name cannot be empty.")
            except ValueError:
                print("Marks must be a number.")

        # Display all students
        elif choice == 2:
            if student_data:
                print("Student Records:")
                for name, marks in student_data.items():
                    print(f"{name} : {marks}")
            else:
                print("No student records found.")

        # Average marks
        elif choice == 3:
            if student_data:
                avg = sum(student_data.values()) / len(student_data)
                print(f"Average Marks: {avg:.2f}")
            else:
                print("No data to calculate average.")

        # Top student
        elif choice == 4:
            if student_data:
                top_student = max(student_data, key=student_data.get)
                print(f"Top Student: {top_student} with {student_data[top_student]} marks")
            else:
                print("No student data available.")

        # Exit
        elif choice == 5:
            print(f"Thank you {name} !...")
            break

        else:
            print("Invalid choice. Please try again.")


students()
```
